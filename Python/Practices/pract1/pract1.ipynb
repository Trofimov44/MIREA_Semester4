Существует старинный метод умножения по методу русского крестьянина. Разобрать его проще на примере.

Предположим, мы хотим перемножить 10 и 15:

| x    | y        |
|------|----------|
| 10 | 15 |
| **5**  | 30     |
| 2      | 60     |
| **1**  | 120    |

В первом столбце последовательно записывают результаты деления на 2 с отбрасыванием остатка. Во втором столбце находятся результаты умножения на 2. Отмечаем нечетные числа в первом столбце. Складываем те числа в правом столбце, которые стоят напротив отмеченных ранее чисел. То есть 30 + 120 = 150. В нашем случае мы не учитывали исходные числа при сложении, но учитывать их, вообще говоря, может быть нужно.

В этом алгоритме используются лишь простейшие операции: умножение на 2, целочисленное деление на 2, проверка на нечетность и сложение. Эти операции соответствуют тем элементарным действиям, которые эффективно выполняет любой процессор. Сам же алгоритм, несмотря на кажущуюся необычность, является вариантом умножения в столбик при использовании двоичного представления чисел:

```
   1111
   1010 *
   ----
   0000
  1111
 0000
1111
```

**3.5.** (уровень сложности: средний)

Реализуйте функцию fast_mul в соответствии с алгоритмом двоичного умножения в столбик (без рекурсии!). Добавьте автоматическое тестирование, как в случае с naive_mul.

````python
def fast_mul(x, y):
    result = 0
    while y > 0:
        if y & 1:  # Если младший бит y равен 1
            result += x  # Добавляем x к результату

        x <<= 1  # Удваиваем x (сдвигаем влево на 1 бит)
        y >>= 1  # Делим y на 2 (сдвигаем вправо на 1 бит)

    return result

print(fast_mul(2, 4))
````

**3.6.** (уровень сложности: средний)

Реализуйте аналогичную функцию fast_pow для возведения в степень. Решение необходимо получить только с помощью небольших модификаций предыдущего решения.

````python
def fast_pow(x, y):
    result = 1
    while y > 0:
        if y & 1:
            result *= x  # Умножаем результат на x

        x *= x
        y >>= 1  # Сдвигаем y вправо на 1 бит (делим y на 2)

    return result

print(fast_pow(2, 3))
````

**3.7.** (уровень сложности: средний)

Предположим, что стоит задача перемножить два 16-битных числа без знака. Однако в наличии у нас имеется только операция умножения 8-битных чисел. Оказывается, можно применить четыре 8-битных умножения, если разбить исходные числа на 8-битные половинки. Побитовые сдвиги умножениями не считаются.

Выведите формулу для умножения «половинок» и реализуйте функцию mul16(x, y), используя приведенную ниже функцию mul_bits, которая имитирует умножение с ограниченной числом bits разрядностью аргументов.

````python
def mul_bits(x, y, bits):
    x &= (2 ** bits - 1)
    y &= (2 ** bits - 1)
    return x * y
````

````python
def mul16(x, y):
    x0 = x & 0xFF
    x1 = (x >> 8) & 0xFF

    y0 = y & 0xFF
    y1 = (y >> 8) & 0xFF

    z0 = mul_bits(x0, y0, 8)
    z1 = mul_bits(x1, y1, 8)
    z2 = mul_bits(x1, y0, 8) + mul_bits(x0, y1, 8)

    result = (z1 << 16) + (z2 << 8) + z0
    return result

print(mul16(3, 4))
````

**3.8.** (уровень сложности: средний)

Четыре умножения из предыдущей задачи – не предел. Советский ученый Анатолий Алексеевич Карацуба в 1960 г. [предложил](https://www.mathnet.ru/links/5c522538012c746223cf85fe9ea8fc07/tm1120.pdf) формулу (а для общего случая – рекурсивный алгоритм), требующую лишь трех умножений. Выведите эту формулу с подсказками от преподавателя и реализуйте функцию mul16k(x, y).

````python
def mul16k(x, y):
    x0 = x & 0xFF
    x1 = (x >> 8) & 0xFF

    y0 = y & 0xFF
    y1 = (y >> 8) & 0xFF

    z0 = x0 * y0
    z1 = x1 * y1
    z2 = (x1 + x0) * (y1 + y0) - z1 - z0

    result = (z1 << 16) + (z2 << 8) + z0
    return result
print(mul16k(3, 4))
````

**3.9.** (уровень сложности: высокий)

Иногда возникает необходимость в создании программы, которая в качестве результата выдаст другую программу. В этом нет ничего необычного, так устроен, к примеру, компилятор. Но если задача стоит узкоспециальная, то речь идет о создании генератора программ по заданным параметрам.

Реализуйте генератор программ fast_mul_gen(y) для задач 2.1-2.3. Воспользуйтесь ранее полученным кодом fast_mul. Ваша функция должна выдать текст функции f(x) (умножение на ранее заданный y), тело которой состоит из некоторого числа присваиваний. Для вывода функции используйте print. Добавьте автоматическое тестирование. Объясните, почему в общем случае у вас получается большее количество сложений, чем в задачах 2.1-2.3.

````python
def fast_mul_wrap(x, y):
    result = 0
    while x:
        # Если x нечетное, добавляем y к результату
        if x % 2 != 0:
            result = y + result

        # Удваиваем y и делим x на 2
        x = x // 2  # Делим x на 2 (сдвиг вправо)
        y = y * 2  # Удваиваем y (сдвиг влево)

    return result

print(fast_mul_wrap(1, 12))  # Вызов fast_mul_wrap с y = 12

print(fast_mul_wrap(1, 16))  # Вызов fast_mul_wrap с y = 16

print(fast_mul_wrap(1, 15))  # Вызов fast_mul_wrap с y = 15
````

````python
def fast_pow_wrap(x, y):
    result = 1
    while y > 0:
        if y & 1:
            result *= x
        x *= x
        y >>= 1  # fast_gen
    return result

print(fast_pow_wrap(2, 2))

print(fast_pow_wrap(2, 4))

print(fast_pow_wrap(2, 8))
````

Изобразите шар, показанный на примере ниже.

![](images/pyshader1.png)

````python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

# Создаем и рисуем изображение
width, height = 256, 256
image = bytearray((0, 0, 0) * width * height)

# Функция шейдера для круга
for y in range(height):
    for x in range(width):
        # Нормализуем координаты
        x_norm = x / width
        y_norm = y / height

        s = 0.6
        d = 0.53
        d2 = 1 - d
        r2 = ((x_norm - d2) ** 2 + (y_norm - d2) ** 2) * 2 - (s ** 2)
        r1 = ((x_norm - d) ** 2 + (y_norm - d) ** 2) * 2 - (s ** 2)

        # Цвет (R, G, B) - нормализуем значения в диапазон 0-1
        r = -r1 - (r1 * 2)  # красный канал
        g = -r2 + (-r2 * 2)  # зеленый канал
        b = 0  # синий канал

        # Нормализуем цвета в диапазон 0-255
        normalized = [max(min(int(c * 255), 255), 0) for c in [r, g, b]]
        pos = (width * y + x) * 3
        image[pos:pos + 3] = normalized

# Преобразуем байты в изображение PIL
img = Image.frombytes('RGB', (width, height), bytes(image))

# Увеличиваем изображение в 2 раза (аналог zoom(2,2))
img = img.resize((width*2, height*2), Image.NEAREST)

# Отображаем с помощью matplotlib
plt.figure(figsize=(8, 8))
plt.imshow(img)
plt.axis('off')  # Отключаем оси
plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt

def noise(x, y):
    """
    Генерирует псевдослучайное значение на основе координат x, y
    Используем простой хэш на основе координат для создания шума
    """
    n = int(x * 12345 + y * 6789)  # Умножаем координаты на разные простые числа
    n = (n << 13) ^ n  # Побитовая операция для смешивания
    n = (n * (n * n * 15731 + 789221) + 1376312589) & 0x7fffffff  # Простая хэш-функция
    return n / 0x7fffffff  # Нормализуем к диапазону [0, 1]

def interpolate(a, b, t):
    #Линейная интерполяция между a и b с параметром t
    return a + t * (b - a)

def smoothstep(t):
    #Функция сглаживания для более плавного перехода

    return t * t * (3 - 2 * t)

def val_noise(x, y):
    #Генерация интерполяционного шума (value noise) на основе функции noise

    # Определяем координаты узлов сетки
    x0 = int(x)
    x1 = x0 + 1
    y0 = int(y)
    y1 = y0 + 1

    # Получаем значения шума в узлах сетки
    n00 = noise(x0, y0)
    n01 = noise(x0, y1)
    n10 = noise(x1, y0)
    n11 = noise(x1, y1)

    # Вычисляем дробные части координат для интерполяции
    tx = x - x0
    ty = y - y0

    # Применяем функцию сглаживания
    sx = smoothstep(tx)
    sy = smoothstep(ty)

    # Интерполируем значения шума
    nx0 = interpolate(n00, n10, sx)
    nx1 = interpolate(n01, n11, sx)
    return interpolate(nx0, nx1, sy)

# Создаем изображение шума размером 256x256
width, height = 256, 256
noise_image = np.zeros((height, width))

# Заполняем изображение шумом
for y in range(height):
    for x in range(width):
        # Нормализуем координаты и применяем val_noise
        noise_image[y, x] = val_noise(x / 32, y / 32)  # Масштабируем координаты для получения более крупного шума

# Визуализируем шум как градиент серого цвета
plt.figure(figsize=(8, 8))
plt.imshow(noise_image, cmap='gray')
plt.axis('off')  # Убираем оси
plt.show()
````

**4.6.** (уровень сложности: средний)

Изобразите облака с помощью алгоритма фрактального шума (fBm noise, см. источники выше) и с использованием ранее разработанной функции val_noise.

![](images/pyshader6.png)

````python
import numpy as np
import matplotlib.pyplot as plt

def noise(x, y):
    #Простая функция шума, генерирующая псевдослучайное значение на основе координат.
    n = int(x * 12345 + y * 6789)  # Умножаем координаты на разные простые числа
    n = (n << 13) ^ n  # Побитовая операция для смешивания
    n = (n * (n * n * 15731 + 789221) + 1376312589) & 0x7fffffff  # Простая хэш-функция
    return n / 0x7fffffff  # Нормализуем к диапазону [0, 1]

def interpolate(a, b, t):
    #Линейная интерполяция между значениями a и b с весом t (0 ≤ t ≤ 1).
    return a + t * (b - a)

def val_noise(x, y):

    #Функция value noise с линейной интерполяцией.
    x0 = int(x)
    x1 = x0 + 1
    y0 = int(y)
    y1 = y0 + 1

    tx = x - x0
    ty = y - y0

    n00 = noise(x0, y0)
    n01 = noise(x0, y1)
    n10 = noise(x1, y0)
    n11 = noise(x1, y1)

    ix0 = interpolate(n00, n10, tx)
    ix1 = interpolate(n01, n11, tx)

    return interpolate(ix0, ix1, ty)

def fbm_noise(x, y, octaves=6, lacunarity=2.0, gain=0.5):
    #Функция фрактального шума (fBm) для создания облаков.

    value = 0.0
    amplitude = 1.0
    frequency = 1.0

    for _ in range(octaves):
        value += val_noise(x * frequency, y * frequency) * amplitude
        frequency *= lacunarity  # Увеличиваем частоту
        amplitude *= gain  # Уменьшаем амплитуду

    # Нормализуем результат в диапазон [0, 1]
    return max(0.0, min(1.0, value * 0.5 + 0.5))

# Создаем изображение облаков размером 512x512
width, height = 512, 512
cloud_image = np.zeros((height, width))

# Заполняем изображение fBm шумом для облаков
for y in range(height):
    for x in range(width):
        cloud_image[y, x] = fbm_noise(x * 0.02, y * 0.02)

# Применяем цветовую палитру для облаков: белые/светло-голубые облака на голубом небе
colored_clouds = np.zeros((height, width, 3))
for y in range(height):
    for x in range(width):
        value = cloud_image[y, x]
        # Создаем цвет: светло-голубые/белые облака на голубом небе
        r = max(0.6, 1.0 - value * 0.3)  # немного красного для мягкого белого/голубого
        g = max(0.8, 1.0 - value * 0.2)  # больше зеленого для голубого оттенка
        b = max(0.9, 1.0 - value * 0.1)  # еще больше синего для неба и облаков
        colored_clouds[y, x] = [r, g, b]

# Визуализируем облака
plt.figure(figsize=(10, 10))
plt.imshow(colored_clouds)
plt.axis('off')  # Убираем оси
plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt

# Функция, вычисляющая расстояние от точки (x, y) до окружности радиуса r
def circle(x, y, r):
    return (x ** 2 + y ** 2) ** 0.5 - r  # Формула расстояния до окружности

# Функция, вычисляющая расстояние от точки (x, y) до прямоугольника с размерами size_x и size_y
def box(x, y, size_x, size_y):
    dx = abs(x) - size_x
    dy = abs(y) - size_y
    outside_dist = (max(dx, 0)**2 + max(dy, 0)**2) ** 0.5
    inside_dist = min(max(dx, dy), 0)
    return outside_dist + inside_dist  # Общее расстояние

# Функция, задающая SDF (Signed Distance Function) для комбинированной фигуры
def sdf_func(x, y):
    rect = box(x, y, 0.1, 0.4)

    # Внешняя окружность с радиусом 0.3 и смещением (0.2, -0.1)
    outer_circle = circle(x - 0.2, y + 0.1, 0.3)
    # Внутренняя окружность с радиусом 0.1 и тем же смещением
    inner_circle = circle(x - 0.2, y + 0.1, 0.1)
    # Создаем окружность с отверстием (разность между внешней и внутренней окружностью)
    circle_with_hole = max(outer_circle, -inner_circle)

    # Объединяем прямоугольник и окружность с отверстием (логическое ИЛИ)
    return min(rect, circle_with_hole)

# Функция шейдера, которая определяет цвет пикселя на основе расстояния
def shader(x, y):
    d = sdf_func(x - 0.5, y - 0.5)
    return d > 0, abs(d) * 3, 0

# Размеры изображения
width, height = 256, 256
image = np.zeros((height, width, 3))

# Проходим по каждому пикселю изображения
for y in range(height):
    for x in range(width):
        r, g, b = shader(x / width, y / height)
        image[y, x] = [r, g, b]

plt.imshow(image)
plt.axis("off")
plt.show()
````
